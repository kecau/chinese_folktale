```python
import pandas as pd
import numpy as np
```

```python
filename_cn="./data/charactor normolize.csv"
filename_on="./data/original normolize.csv"
filename_sn="./data/space normolize.csv"
filename_tn="./data/time normolize.csv"
filename_result="./data/result.xlsx"
filename_story="./data/story.xlsx"
```

```python
df_story=pd.read_excel(filename_story)
story_map={x:y for x,y in zip(df_story.id,df_story.title)}
story_map_inv={x:y for y,x in story_map.items()}
```

```python
def process(filename):
    df=pd.read_csv(filename)
    df["Unnamed: 0"]=df["Unnamed: 0"].map(story_map_inv)
    df=df.rename(columns=story_map_inv)
    df=df.rename(columns={"Unnamed: 0":"id"})
    df=df.set_index("id")
    newcol=[i for i in range(1,177)]
    df=df[newcol]
    df=df.sort_index()
    return df
```

```python
df_cn=process(filename_cn)
df_on=process(filename_on)
df_sn=process(filename_sn)
df_tn=process(filename_tn)
df_re=pd.read_excel(filename_result)
```

```python
rename_map={'original similarity result1(A,B)':"on(A,B)",'original similarity result2(B,C)':"on(B,C)",
       'original similarity result3(A,C)':"on(A,C)", 'charactor wildcard1(A,B)':"cn(A,B)",
       'charactor wildcard2(B,C)':"cn(B,C)", 'charactor wildcard3(A,C)':"cn(A,C)",
       'time wildcard1(A,B)':"tn(A,B)", 'time wildcard2(B,C)':"tn(B,C)", 'time wildcard3(A,C)':"tn(A,C)",
       'space wildcard1(A,B)':"sn(A,B)", 'space wildcard2(B,C)':"sn(B,C)", 'space wildcard3(A,C)':"sn(A,C)"}
df_re=df_re.rename(columns=rename_map)
```

```python
def generateshunxi(i):
    shunxi=[[],[],[]]
    need=[]
    maxcount= max(df_re.iloc[i,:]["A,B"],df_re.iloc[i,:]["B,C"],df_re.iloc[i,:]["A,C"])
    for k in ["A,B","A,C","B,C"]:
        if maxcount==df_re.iloc[i,:][k]:
            shunxi[0].append((k))
    if len(shunxi[0])==2:
        for k in ["A,B","A,C","B,C"]: 
            if maxcount!=df_re.iloc[i,:][k]:
                shunxi[1].append((k))
    elif len(shunxi[0])==3:
        pass
    else:  # ==1
        for k in ["A,B","A,C","B,C"]: 
            if k not in shunxi[0]:
                need.append(k)
        if df_re.iloc[i,:][need[0]]>df_re.iloc[i,:][need[1]]:
            shunxi[1].append((need[0]))
            shunxi[2].append((need[1]))
        elif df_re.iloc[i,:][need[0]]==df_re.iloc[i,:][need[1]]:
            shunxi[1].append((need[0]))
            shunxi[1].append((need[1]))
        else:
            shunxi[1].append((need[1]))
            shunxi[2].append((need[0]))
    return shunxi
```

```python
for i in range(250):
    shunxi=generateshunxi(i)
    # if shunxi[2]:
    #     print(i,generateshunxi(i))
    if len(shunxi[0])==2:
        print(i,generateshunxi(i))
```

```python
def out(df_re,col):
    acc=0
    mrr=0
    for i in range(len(df_re)):
        shunxi=generateshunxi(i)
        temp=0
        need=[]
        maxrel= max(df_re.iloc[i,:][col+"("+"A,B"+")"],df_re.iloc[i,:][col+"("+"B,C"+")"],df_re.iloc[i,:][col+"("+"A,C"+")"])
        for k in ["A,B","A,C","B,C"]:  
            if df_re.iloc[i,:][col+"("+k+")"]==maxrel:
                if k in shunxi[0]:
                    acc+=1
                    temp+=1
        if temp==0:
            for k in ["A,B","A,C","B,C"]:  
                if df_re.iloc[i,:][col+"("+k+")"]!=maxrel:
                    need.append(k)
            if df_re.iloc[i,:][col+"("+need[0]+")"]>df_re.iloc[i,:][col+"("+need[1]+")"]:
                if len(shunxi[0])==2:
                    if need[0] in shunxi[0]:
                        temp+=0.5
                else:
                    if need[0] in shunxi[1]:
                        temp+=0.5
                    else:
                        if need[1] in shunxi[2]:
                            temp+=1/3
            else:
                if len(shunxi[0])==2:
                    if need[1] in shunxi[0]:
                        temp+=0.5
                else:
                    if need[1] in shunxi[1]:
                        temp+=0.5
                    else:
                        if need[0] in shunxi[2]:
                            temp+=1/3
            # if shunxi[2]:
            #     print(shunxi[2],df_re.iloc[i,:][col+"("+need[0]+")"],df_re.iloc[i,:][col+"("+need[1]+")"])
        assert temp<=1
        # print(temp)
        # print(acc)
        mrr+=temp
    return acc/len(df_re),mrr/len(df_re)
```

```python
acc_on,mrr_on=out(df_re,"on")
acc_tn,mrr_tn=out(df_re,"tn")
acc_sn,mrr_sn=out(df_re,"sn")
acc_cn,mrr_cn=out(df_re,"cn")

```
